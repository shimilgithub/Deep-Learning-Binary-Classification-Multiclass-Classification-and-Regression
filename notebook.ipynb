# Deep Learning Tasks: Binary Classification, Multiclass Classification, and Regression


**This notebook demonstrates the implementation of the classes: Imdb, Reuters, and BostonHousing.**


Each class includes functions to prepare data, build a model, train, plot loss and accuracy, evaluate the model, and save it.

```python
#Import required libraries 
from imdb import Imdb
from reuters import Reuters
from boston_housing import BostonHousing
```

## 1. IMDB - Binary Classification

```python
# Binary Classification - IMDB

#define an instance
imdb_model = Imdb()

#prepare data
imdb_model.prepare_data()

#build model
imdb_model.build_model()

#train model
imdb_model.train()

#plot loss and accuracy
imdb_model.plot_loss()
imdb_model.plot_accuracy()

#Evaluate model
imdb_results = imdb_model.evaluate()
print(f"IMDB Test Loss: {imdb_results[0]}, Test Accuracy: {imdb_results[1]}")

```

**Interpretation:**

The model achieved a test loss of 0.313 which is a relatively low error, and a test accuracy of 87.68%. It shoes  strong performance in correctly predicting on the IMDB dataset.

## 2. Reuters - Multiclass Classification

```python
# Multiclass Classification - Reuters

#define an instance
reuters_model = Reuters()

#prepare data
reuters_model.prepare_data()

#build model
reuters_model.build_model()

#train model
reuters_model.train()

#plot loss and accuracy
reuters_model.plot_loss()
reuters_model.plot_accuracy()

#Evaluate model
reuters_results = reuters_model.evaluate()
print(f"Reuters Test Loss: {reuters_results[0]}, Test Accuracy: {reuters_results[1]}")

```

**Interpretation:**

The model achieved a test loss of 1.089 which indicates moderate error, and a test accuracy of 78.41%. This is a good performance in classifying Reuters news topics.

## 3. Boston Housing - Regression

```python
# Regression - Boston Housing

#define an instance
boston_model = BostonHousing()

#prepare data
boston_model.prepare_data()

#build model
boston_model.build_model()

#train model
boston_model.train()

#plot loss and MAE and MSE/RMSE
boston_model.plot_loss()
boston_model.plot_mae()
boston_model.calculate_mse_rmse()

#Evaluate model
boston_results = boston_model.evaluate()
print(f"Boston Housing Test Loss: {boston_results[0]}, Test MAE: {boston_results[0]}")
```

**Interpretation:**

The model's performance on the Boston Housing dataset shows a Mean Squared Error (MSE) of 17.1, a Root Mean Squared Error (RMSE) of 4.1, and a Mean Absolute Error (MAE) of 16.86. This is a good performance indication.
